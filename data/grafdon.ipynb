```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cnt = 2
df = pd.read_csv(f"maru{cnt}.csv", sep=",")
```

```python
c1,c2,c3,c4,c5,c6 = "blue","green","red","black" ,"yellow","orange"     # 各プロットの色
l1,l2,l3,l4,l5,l6 = "Gx[dps]","Gy[dps]","Gz[dps]","Ax[g]", "Ay[g]", "Az[g]"   # 各ラベル
```

```python
fig, ax = plt.subplots()

plt.plot(df["time[ms]"], df["Gx[mg]"], color=c1, label=l1)
plt.plot(df["time[ms]"], df["Gy[mg]"], color=c2, label=l2)
plt.plot(df["time[ms]"], df["Gz[mg]"], color=c3, label=l3)

ax.grid()
ax.legend(loc=0)
ax.set_xlabel("time[ms]")
ax.set_ylabel("gyro[mg]")
fig.tight_layout()
plt.show()
fig.savefig(f"marug{cnt}.png")
```

```python
fig, ax = plt.subplots()

plt.plot(df["time[ms]"], df["Ax[dps]"], color=c4, label=l4)
plt.plot(df["time[ms]"], df["Ay[dps]"], color=c5, label=l5)
plt.plot(df["time[ms]"], df["Az[dps]"], color=c6, label=l6)

ax.grid()
ax.legend(loc=0)
ax.set_xlabel("time[ms]")
ax.set_ylabel("acc[dps]")
fig.tight_layout()
plt.show()
fig.savefig(f"marua{cnt}.png")
```
